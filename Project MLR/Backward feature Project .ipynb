```python
X = df.drop(columns=['Id','SalePrice'])
Y = df[['SalePrice']]
```

```python
X.columns
```

```python
X.dtypes
```

```python
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1  = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                            ('scaler',StandardScaler())])
```

```python
cat_pipe1 = Pipeline(steps=[('impute',SimpleImputer(strategy='constant',fill_value='Not_Avail')),
                           ('ordinal',OrdinalEncoder())])
```

```python
pre1 = ColumnTransformer([('num',num_pipe1,con),
                         ('cat',cat_pipe1,cat)]).set_output(transform='pandas')
```

```python
pre1
```

```python
X_pre = pre1.fit_transform(X)
```

```python
X_pre.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
lr = LinearRegression()
sel = SequentialFeatureSelector(estimator=lr,n_features_to_select='auto',direction='backward')
sel.fit_transform(X_pre,Y)
sel_cols = sel.get_feature_names_out()
sel_cols
```

```python
len(sel_cols)
```

```python
sel_cols[0]
```

```python
sel_cols[0].split('__')
```

```python
imp_cols = []
for i in sel_cols:
    s = i.split('__')[1]
    imp_cols.append(s)
```

```python
imp_cols
```

```python
X_sel = X[imp_cols]
X_sel.head()
```

```python
cat_sel = list(X_sel.columns[X_sel.dtypes=='object'])
con_sel = list(X_sel.columns[X_sel.dtypes!='object'])
```

```python
cat_sel
```

```python
con_sel
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = Pipeline(steps=[('impute',SimpleImputer(strategy='mean')),
                            ('scaler',StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('impute',SimpleImputer(strategy='constant',fill_value='Not_Avail')),
                            ('ohe',OneHotEncoder(handle_unknown='ignore',sparse_output=False))])
```

```python
pre2 = ColumnTransformer([('num',num_pipe2,con_sel),
                          ('cat',cat_pipe2,cat_sel)]).set_output(transform='pandas')
```

```python
pre2
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X_sel_pre, Y ,test_size=0.2,random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
model1 = LinearRegression()
model1.fit(xtrain,ytrain)
```

```python
model1.score(xtrain,ytrain)
```

```python
model1.score(xtest,ytest)
```

```python
from sklearn.linear_model import Ridge 
model2 = Ridge(alpha=1)
model2.fit(xtrain, ytrain)
```

```python
model2.score(xtrain,ytrain)
```

```python
model2.score(xtest,ytest)
```

```python
def evalute_model(x,y):
    ypred = model2.predict(x)
    from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error,r2_score
    mse = mean_squared_error(y,ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y,ypred)
    mape = mean_absolute_percentage_error(y,ypred)
    r2 = r2_score(y,ypred)
    print(f'MSE:{mse:.2f}')
    print(f'RMSE:{rmse:.2f}')
    print(f'MAE :{mae:.2f}')
    print(f'MAPE : {mape:.4f}')
    print(f'R2:{r2:.4f}')
```

```python
evalute_model(xtrain,ytrain)
```

```python
evalute_model(xtest,ytest)
```

```python
from sklearn.linear_model import Lasso
model3 = Lasso(alpha=1)
model3.fit(xtrain,ytrain)
```

```python
model3.score(xtrain,ytrain)
```

```python
model3.score(xtest,ytest)
```

```python
def evalute_model(x,y):
    ypred = model3.predict(x)
    from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error,r2_score
    mse = mean_squared_error(y,ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y,ypred)
    mape = mean_absolute_percentage_error(y,ypred)
    r2 = r2_score(y,ypred)
    print(f'MSE:{mse:.2f}')
    print(f'RMSE:{rmse:.2f}')
    print(f'MAE :{mae:.2f}')
    print(f'MAPE : {mape:.4f}')
    print(f'R2:{r2:.4f}')
```

```python
evalute_model(xtrain,ytrain)
```

```python
evalute_model(xtest,ytest)
```
